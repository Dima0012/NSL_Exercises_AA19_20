In questa esercitazione andremo a confrontare il prezzo di una opzione europea *call* e *put* stimato tramite metodi Monte Carlo, con il valore esatto ricavato dalla teoria di Black-Scholes.

### Esercizio 03.01
Vogliamo calcolare il prezzo di acquisto (*call*, $C[S(t),t]$) e vendita (*put*, $P[S(t),t)$) di una opzione europea al tempo iniziale $t=0$, dato un prezzo di asset $S(t)$. La teoria di Black-Scholes fornisce una soluzione *analitica* (quindi esatta) al problema, sotto alcune ipotesi:

1. Il mercato deve essere massimamente efficente.
2. L'evoluzione temporale del prezzo di asset è stocastica ed esibisce un moto Browniano geometrico, con tasso di interesse senza rischio $r$ e volatilità $\sigma$ costanti: $GBM(r, \sigma^2)\,$.

Fissiamo i seguenti parametri, che definiscono il problema del prezzo dell'opzione e rientrano esplicitamente nella soluzione analitica, assegnando anche dei valori numerici:

- Prezzo dell'asset a $t=0$: $S(0)=100$
- Data di estinzione: $T=1$
- Prezzo di esercizio: $K=100$
- Tasso di interesse senza rischio: $r=0.1$
- Volatilità: $\sigma=0.25$

Con questi valori, la soluzione esatta del problema è:

$$Call: \quad C[S(0),0] = 14.9758  \qquad \qquad   Put: \quad P[S(0),0) = 5.4595 $$

E' possibile approcciare il problema sfruttando metodi Monte Carlo, quando le ipotesi di Black-Scholes vengono a mancare; assumiamo che il prezzo di asset evolve in modo assimilabile ad un moto Browniano: $S\sim GBM(r, \sigma^2)\,$.

Si ha:

$$C[S(0),0] = e^{-rT}max(\,0, S(T) - K \,)  \qquad \qquad  P[S(0),0) = e^{-rT}max(\,0, K - S(T) \,) $$

Il valore di $S(T)$ (prezzo dell'asset alla scadenza) si ottiene tramite un campionamento da moto Browniano; in particolare è possibile fare un campionamento *diretto* oppure *discretizzato*: 

$$Diretto: \quad S(T) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)T+\sigma W(T)\sqrt{T}\right]\,,  \quad W(T)\sim\mathcal{N}(0,T)$$

$$Discretizzato: \quad S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i) +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right] \,, \quad Z_i \sim \mathcal{N}(0,1)$$

in cui $\,0\equiv t_0 < t_1 < \dots < t_n \equiv T \,$ sono i tempi discreti in cui si suddivide l'intervallo $[0, T]\,$.

Procediamo con entrambi i metodi; ripetiamo il campionamento un numero $N_{step}=10^4$ di volte, che suddividiamo in $N_{blocks}=100$ blocchi in modo da potere utilizzare il metodo a blocchi per ricavare valore medio ed errore. I risultati (confrontati con il valore esatto) sono riportati di seguito.

```python
# Moduli
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
fig1, ( (ax0, ax1), (ax2, ax3) ) = plt.subplots(nrows=2, ncols=2, figsize=(15,15))

n_step = 1e4
n_blocks = 100 

x = np.arange(n_blocks)
mean_prog = np.zeros(n_blocks)
err_prog = np.zeros(n_blocks)

expected_call = [14.975790778311286 for i in x] #Valore atteso Call
expected_put = [5.4595325819072364 for i in x]  #Valore atteso Put

#################################

#Call Diretta

file_mean = open("03.1/Data/call_1.out") # Dati della somma progressiva
file_err = open("03.1/Data/call_1_err.out") # Dati dell'errore progressivo
    
for i in range(n_blocks):
    mean_prog[i] = float(file_mean.readline())
    err_prog[i] = float(file_err.readline())
    
file_mean.close()
file_err.close()
    
    
ax0.errorbar(x,mean_prog,yerr=err_prog)
ax0.set_xlabel('# Blocchi', fontsize='large')
ax0.set_ylabel('$C[S(0), 0]$', fontsize='large')
ax0.set_title('$C\,[S(0), 0]}$ con samplig diretto', fontsize='x-large')
ax0.plot(x,expected_call, 'r', label='Call Attesa')
ax0.grid(True, linestyle="--", linewidth=0.2, color='.25', zorder=-10)

###################################

#Put Diretta

mean_prog = np.zeros(n_blocks)
err_prog = np.zeros(n_blocks)


file_mean = open("03.1/Data/put_1.out") # Dati della somma progressiva
file_err = open("03.1/Data/put_1_err.out") # Dati dell'errore progressivo
    
for i in range(n_blocks):
    mean_prog[i] = float(file_mean.readline())
    err_prog[i] = float(file_err.readline())
    
file_mean.close()
file_err.close()
    
    
ax1.errorbar(x,mean_prog,yerr=err_prog)
ax1.set_xlabel('# Blocchi', fontsize='large')
ax1.set_ylabel('$P[S(0), 0]$', fontsize='large')
ax1.set_title('$P\,[S(0), 0]}$ con sampling diretto', fontsize='x-large')
ax1.plot(x,expected_put, 'g', label='Put Attesa')
ax1.grid(True, linestyle="--", linewidth=0.2, color='.25', zorder=-10)

###################################

#Call Discretizzata

mean_prog = np.zeros(n_blocks)
err_prog = np.zeros(n_blocks)


file_mean = open("03.1/Data/call_2.out") # Dati della somma progressiva
file_err = open("03.1/Data/call_2_err.out") # Dati dell'errore progressivo
    
for i in range(n_blocks):
    mean_prog[i] = float(file_mean.readline())
    err_prog[i] = float(file_err.readline())
    
file_mean.close()
file_err.close()
    
    
ax2.errorbar(x,mean_prog,yerr=err_prog)
ax2.set_xlabel('# Blocchi', fontsize='large')
ax2.set_ylabel('$C[S(0), 0]$', fontsize='large')
ax2.set_title('$C\,[S(0), 0]}$ con samplig discretizzato', fontsize='x-large')
ax2.plot(x,expected_call, 'r')
ax2.grid(True, linestyle="--", linewidth=0.2, color='.25', zorder=-10)

###################################

#Put Discretizzata

mean_prog = np.zeros(n_blocks)
err_prog = np.zeros(n_blocks)


file_mean = open("03.1/Data/put_2.out") # Dati della somma progressiva
file_err = open("03.1/Data/put_2_err.out") # Dati dell'errore progressivo
    
for i in range(n_blocks):
    mean_prog[i] = float(file_mean.readline())
    err_prog[i] = float(file_err.readline())
    
file_mean.close()
file_err.close()
  
ax3.errorbar(x,mean_prog,yerr=err_prog)

ax3.set_xlabel('# Blocchi', fontsize='large')
ax3.set_ylabel('$P[S(0), 0]$', fontsize='large')
ax3.set_title('$P\,[S(0), 0]$ con samplig discretizzato', fontsize='x-large')
ax3.plot(x,expected_put,'g')
ax3.grid(True, linestyle="--", linewidth=0.2, color='.25', zorder=-10)

fig1.legend(loc='lower center', ncol=2, fontsize='x-large')


plt.show()
```

Vediamo come in tutti i casi il valore calcolato converge al valore atteso, entro una deviazione standard.
