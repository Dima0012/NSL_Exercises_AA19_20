In questa esercitazione andremo a ottimizzare tramite metodi Monte Carlo  variazionali (***VMC***) lo stato fondamentale *GS* di una particella in una dimensione soggetta al potenziale esterno:
$$
V(x) = x^4 - \frac{5}{2}x^2.
$$
Confronteremo i risultati ottenuti con una soluzione ricavata numericamente per l'energia del *GS*, e con un codice che sfrutta il *Path Integral Ground State* (***PIGS***) e il *Path Integral Monte Carlo* (***PIMC***) per sistemi a temperature $T>0$.

### Esercizio 08.1
Il problema con il potenziale dato non è risolubile analiticamente. E' quindi necessario utilizzare metodi numerici per ricavare l'energia del *GS* e la rispettiva funzione d'onda. Utilizziamo 3 procedure distinte:

- Soluzione numerica calcolata diagonalizzando la matrice associata all'equazione di Schrödinger indipendente dal tempo che descrive il problema: gli autovalori e autovettori di questa matrice sono esattamente la funzione d'onda e i valori di energia a cui siamo interessati.
- Utilizzo di un codice *Variational Monte Carlo* (***VMC***) che sfrutta un approccio variazionale, partendo da una funzione trial, per ricavare una stima di funzione d'onda ed energia del *GS*.
- Utilizzo di un codice che sfrutta il *Path Integral Monte Carlo* (***PIMC***), sempre partendo da funzioni d'onda trial.

Di seguito riportiamo il codice Phyton che permette di calcolare la souzione numerica, e i relativi risultati per le energia dei primi tre stati.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])
print("1st excited state energy: ", E[1])
print("2nd excited state energy: ", E[2])

# Plot a few things
#plt.figure(figsize=(8,5))
#scale = 0.3
#plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
#plt.plot(x,(psi[0])**2)
#plt.plot(x,(psi[1])**2)
#plt.plot(x,(psi[2])**2)
#plt.title("Potential & Probabilities")
#plt.xlabel("x")
#plt.grid(True)
#plt.xlim((-3,3))
#plt.ylim((-0.6,0.6))
#plt.show()
```

Il codice ***VMC*** si basa sul principio variazionale, per cui:

$$
\langle {\hat H} \rangle_T = \frac{\langle \Psi_T^{\sigma,\mu}| {\hat H} | \Psi_T^{\sigma,\mu} \rangle}
{\langle \Psi_T^{\sigma,\mu} | \Psi_T^{\sigma,\mu} \rangle} \ge E_0 =
\frac{\langle \Psi_0| {\hat H} | \Psi_0 \rangle}
{\langle \Psi_0 | \Psi_0 \rangle}
$$

dove $E_0$ e $\Psi_0$ sono rispettivamente l'energia e la funzione d'onda dello stato fondamentale. 

$\Psi_T^{\sigma,\mu}$ è invece la funzione d'onda trial che dipende dai due parametri $\sigma$ e $\mu$ ed è definita come:

$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

Il codice fissa un dato set di parametri iniziali per la funzione trial; poi sfrutta l'algoritmo di Metropolis per campionare la distribuzione di probabilità associata a questa funzione d'onda di prova, per un set di parametri iniziali casuali distribuiti uniformemente in un intervallo centrato sui parametri iniziali.

Il codice calcola infine il valore di aspettazione dell'energia $\langle{\hat H}\rangle_T$ per i parametri casuali e lo confronta col il valore precedente a parametri fissati $\langle{\hat H}\rangle_{old}$. Se accade che $\langle{\hat H}\rangle_T < \langle{\hat H}\rangle_{old}$, i valori di $\sigma$ e $\mu$ sono aggiornati e si riduce l'intervallo per la generazione cauale di nuovi parametri, altrimenti rimangono i parametri precedenti.

Questo processo è ripetuto un numero sufficente di volte: alla fine di questo processo di minimizzazione si ottengono dei valori dei parametri che definiscono la funzione trial, e il valore di aspettazione dell'energia; questi dovrebbero essere vicini alle quantità reali.

### Esercizio 08.2
Riportiamo di seguito i risultati ottenuti per il calcolo del valore di aspettazione $\langle{\hat H}\rangle_T$ confrontati con il risultato del calcolo numerico della sezione precedente.

Il codice è partito dai seguenti parametri iniziali:

$$\mu = 1.0 \qquad \sigma = 1.0 $$

e l'ampiezza dell'intervallo in cui generare i nuovi parametri è $2$ per entrambi.

Il processo di minimizzazione ha prodotto i seguenti valori, che definiscono la funzione trial:

$$\mu = 0.820991 \qquad \sigma = 0.647811 $$

Si sono eseguti $500$ tentativi di ottimizzazione; ciascuno è composto da $200$ blocchi di mille *step* Metropolis. Il valore dello step del Metropolis è stato tarato in modo da avere un rate di accettazione della mossa (mediato su i vari tentativi di ottimizzazione) di circa il $50\%$.

I valori riportati si riferiscono a quelli dell'utltima ottimizzazione; valore medio e errore statistico sono stati calcolati tramite il metodo a blocchi.

```python
#<H>_T

fig1, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,15))

nblock, H_mean, H_err = np.loadtxt("08.1/Data/output.etot.0", usecols=(0,2,3), unpack=True)

expected = [ E[0] for i in nblock ]

ax.errorbar(nblock,H_mean, yerr=H_err)
ax.plot(expected, color="r", label="Valore atteso")
ax.set_xlabel('Blocco', fontsize='large')
ax.set_ylabel(r'$<H>_T$', fontsize='large')
ax.set_title(r'Valore minimizzato di $<H>_T$', fontsize='x-large')
ax.grid(True, linestyle="--", linewidth=0.2, color='.25', zorder=-10)
ax.legend(fontsize="x-large")

plt.show()
```

Il valore calcolato con il ***VMC*** ( $\langle{\hat H}\rangle_T = -0.3867 $ ) è molto discostato da quello trovato col metodo numerico.

Riportiamo anche la densità di probabilità della funzione d'onda, campionata tramite il codice; la confrontiamo con quella ottenuta dal calcolo numerico, e con la forma analitica usando i parametri troavati con l'ottimizzazione.

```python
#|Psi|^2

mu = 0.820991
sigma = 0.647811

# Defining wavefuncion

def Psi_T(x):
    a=((x-mu)/sigma)**2
    b=((x+mu)/sigma)**2
    return np.exp(-a/2)+np.exp(-b/2)

fig1, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,15))

#Analytical Psi
N=np.sum(dx*Psi_T(x)**2) #Normalization
ax.plot(x,(Psi_T(x)**2)/N, label="Analitica ottimizzata")


#Numerical solution
ax.plot(x,psi[0]**2, label="Soluzione numerica")

#Sampled
Psi_samp = np.loadtxt("08.1/Data/position.0", unpack=True)
nbins = 200

ax.hist(Psi_samp,nbins,range=(-4,4),label='Campionata',density='true', color="yellow")


ax.set_xlabel('x', fontsize='large')
ax.set_ylabel(r'$\rho(x)$', fontsize='large')
ax.set_title(r'Confronto tra varie $\Psi$ - (1)', fontsize='x-large')
ax.grid(True, linestyle="--", linewidth=0.2, color='.25', zorder=-10)
ax.legend(fontsize="x-large")

ax.set_xlim([-3, 3])

plt.show()
```

Nonostante il valore di energia si discosti da quello numerico, il profilo della densità di probabilità è in buon accordo con quello numerico e analitico.

### Esercizio 08.3
Confrontiamo ora la distribuzione di probabilità campionata attraverso il codice ***VMC*** con altre due distribuzioni, calcolate tramite il *Path Integral Ground State* (***PIGS***), utilizzando come funzioni trial:

- La funzione costante $\Psi_T(x)\equiv 1$
- La stessa funzione del ***VMC***, $\Psi_T(x) = \Psi_T^{\sigma,\mu}(x)$ , utilizzando sempre i parametri ottimizzati.

Di seguito riportiamo i risultati.

```python
#Sampled vs Const vs Trial

fig1, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,15))

x_const, psi_const, err_const = np.loadtxt("08.3/constant/probability.dat", usecols=(0,1,2), unpack=True)
x_trial, psi_trial, err_trial = np.loadtxt("08.3/trial/probability.dat", usecols=(0,1,2), unpack=True)


ax.hist(Psi_samp,nbins,range=(-4,4),label='Campionata',density='true', color="yellow")
ax.errorbar(x_const, psi_const, yerr=err_const, label="PIGS Constante")
ax.errorbar(x_trial, psi_trial, yerr=err_trial, label="PIGS Trial")

ax.set_xlabel('x', fontsize='large')
ax.set_ylabel(r'$\rho(x)$', fontsize='large')
ax.set_title(r'Confronto tra varie $\Psi$ - (2)', fontsize='x-large')
ax.grid(True, linestyle="--", linewidth=0.2, color='.25', zorder=-10)
ax.legend(fontsize="x-large")

ax.set_xlim([-3, 3])

plt.show()
```

Di nuovo, c'è un buon accordo tra tutte le distribuzioni; in particolare le due distribuzioni calcolate tramite ***PIGS*** sono pressochè indistinguibili, salvo che nella regione centrale.

```python
# Various temperature PIMC 

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,15))

x_05, psi_05, err_05 = np.loadtxt("08.3/temperature/Data/probability_05.dat", usecols=(0,1,2), unpack=True)
x_1, psi_1, err_1 = np.loadtxt("08.3/temperature/Data/probability_1.dat", usecols=(0,1,2), unpack=True)
x_2, psi_2, err_2 = np.loadtxt("08.3/temperature/Data/probability_2.dat", usecols=(0,1,2), unpack=True)
x_4, psi_4, err_4 = np.loadtxt("08.3/temperature/Data/probability_4.dat", usecols=(0,1,2), unpack=True)
x_6, psi_6, err_6 = np.loadtxt("08.3/temperature/Data/probability_6.dat", usecols=(0,1,2), unpack=True)
x_8, psi_8, err_8 = np.loadtxt("08.3/temperature/Data/probability_8.dat", usecols=(0,1,2), unpack=True)


ax.errorbar(x_05, psi_05, yerr=err_05, label="T = 0.5")
ax.errorbar(x_1, psi_1, yerr=err_1, label="T = 1.0")
ax.errorbar(x_2, psi_2, yerr=err_2, label="T = 2.0")
ax.errorbar(x_4, psi_4, yerr=err_4, label="T = 4.0")
ax.errorbar(x_6, psi_6, yerr=err_6, label="T = 6.0")
ax.errorbar(x_8, psi_8, yerr=err_8, label="T = 8.0")

ax.plot(x,psi[0]**2, label="Sol. num. GS")
ax.plot(x,psi[1]**2, label="Sol. num. stato eccitato")

ax.set_xlabel('x', fontsize='large')
ax.set_ylabel(r'$\rho(x)$', fontsize='large')
ax.set_title(r'$\Psi$ per varie $T$ con PIMC', fontsize='x-large')
ax.grid(True, linestyle="--", linewidth=0.2, color='.25', zorder=-10)
ax.legend(fontsize="x-large")

ax.set_xlim([-3, 3])

plt.show()
```
