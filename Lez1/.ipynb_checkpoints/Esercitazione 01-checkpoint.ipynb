### Esercizio 01.1
Per verificare la bontà del generatore di numeri pseudo-casuali (*RNG*), testiamo la distribuzione uniforme tra 0 e 1.  Chiamato $r$ il numero estratto ( $r\in[0,1)$ ), valutiamo:

- Il valore medio della distribuzione uniforme tra 0 e 1: $\langle r \rangle = \int_0^1 r dr = 1/2$ .
- Il valore della deviazione standard della media per la quantità $\langle (r-1/2)^2 \rangle$ :  $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$ .

I dati prodotti ($M$ estrazioni di $r$, suddivise in $N$ blocchi) sono elaborati tramite il metodo della media a blocchi per produrre una stima del valore medio e dell'errore (statistico) associato senza correlazioni. I risultati sono mostrati nei grafici che seguono.

```python
# Moduli
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

from scipy.stats import norm
from scipy.stats import cauchy

from IPython.display import display, Math, Latex
```

```python
# <r>

M=1e6       # Lanci Totali
N=100       # Blocchi
L=int(M/N)  

x = np.arange(N)
sum_prog = np.zeros(N)
err_prog = np.zeros(N)

file1 = open("01.1/Data/sum_prog_1.out") # Dati della somma progressiva
file2 = open("01.1/Data/err_prog_1.out") # Dati dell'errore progressivo
    
for i in range(N):
    sum_prog[i] = float(file1.readline())
    err_prog[i] = float(file2.readline())
    
expected = [0.5 for i in range(N)]
    
file1.close()
file2.close()
    
fig0, ax0 = plt.subplots(nrows=1, ncols=1, figsize=(10,10))

ax0.errorbar(x,sum_prog,yerr=err_prog)
ax0.plot(x, expected, 'r', label="Valore atteso")

ax0.set_xlabel('# Blocchi', fontsize='large')
ax0.set_ylabel(r'$\langle r \rangle$', fontsize='large')
ax0.set_title(r"Stima di $\langle r \rangle$", fontsize='x-large')
ax0.grid(True, linestyle="--", linewidth=0.2, color='.25', zorder=-10)
ax0.legend(fontsize="large")

plt.show()
```

```python
#sigma^2

sum_prog = np.zeros(N)
err_prog = np.zeros(N)

file1 = open("01.1/Data/sum_prog_2.out")
file2 = open("01.1/Data/err_prog_2.out")
    
for i in range(N):
    sum_prog[i] = float(file1.readline())
    err_prog[i] = float(file2.readline())
    
file1.close()
file2.close()
    
fig0, ax0 = plt.subplots(nrows=1, ncols=1, figsize=(10,10))

expected = [1./12 for i in range(N)]

ax0.errorbar(x,sum_prog,yerr=err_prog)
ax0.plot(x, expected, 'r', label="Valore atteso")

ax0.set_xlabel('# Blocchi', fontsize='large')
ax0.set_ylabel(r'$\sigma^2$', fontsize='large')
ax0.set_title(r"Stima di $\sigma^2$", fontsize='x-large')
ax0.grid(True, linestyle="--", linewidth=0.2, color='.25', zorder=-10)
ax0.legend(fontsize="large")

plt.show()  
```

In entrambi i casi, vediamo come il valore stimato tende a convergere sul valore esatto al crescere dei blocchi, entro una deviazione standard della media.

Un'ulteriore verifica che possiamo effettuare è sfruttare il test $\chi^2$ per stimare la probabilità che i numeri estratti ***non*** provengano da una distribuzione uniforme; se questa probabilità è piccola possiamo conludere che il **RNG** funziona correttamente.

Se dividiamo $[0,1]$ in $a$ sottointervalli ed estraiamo $n$ numeri casuali, il numero di eventi osservati in ogni sottointervallo è $n/a$; implementiamo il $\chi^2$ come:

$$\chi^2 = \sum_{i=1}^a \frac{\left( n_i - n/a \right)^2}{n/a}$$

Fissati $a=100$ e $n=10^4$, possiamo effettuare il test su $10^4$ numeri estratti ogni volta,ripetendolo $100$ volte. In questo caso ci aspettiamo che in media il valore del $\chi^2$ sia circa $100$, il numero di sottointervalli.

```python
#Chi2

a = 100  # Sottointervalli

chi_2 = np.zeros(a)
x = np.arange(a)

file = open("01.1/Data/chi_2.out")  #Dati Chi2

for i in range(a):
    chi_2[i] = float(file.readline())
    
file.close()

fig0, ax0 = plt.subplots(nrows=1, ncols=1, figsize=(10,10))

expected = [100 for i in range(a)]

ax0.scatter(x,chi_2, color="b")
ax0.plot(x,expected,'r',label="Valore atteso")

ax0.set_xlabel('Ripetizioni test', fontsize='large')
ax0.set_ylabel(r'$\chi^2$', fontsize='large')
ax0.set_title(r"Calcolo del $\chi^2$", fontsize='x-large')
ax0.grid(True, linestyle="--", linewidth=0.2, color='.25', zorder=-10)
ax0.legend(fontsize="large")

plt.show()
```

Si vede dal grafico come i risultati del test sono distribuiti attorno al valore atteso; possiamo quindi concludere che la probabilità che la distribuzione non sia uniforme è piccola e affermare che il **RNG** funziona correttamente.

- Distribuzione uniforme (dado a 100 facce).

### Esercizio 01.3
Ci proponiamo di simulare l'esperimento di Buffon per stimare il valore di $\pi$:
un ago di lunghezza $L$ è lasciato cadere in modo casuale su un piano orizzontale cosparso di righe parallele ed equispaziate a distanza $d>L$: la probabilità che l'ago intersechi una linea è $P = 2L/\pi d$ . Si può allora stimare il valore di $\pi$ tramite:

$$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d},$$

dove $N_{hit}$ è il numero di volte che l'ago interseca una linea, per un totale di $N_{thr}$ lanci. Elaboriamo i dati ricavati tramite il metodo a blocchi.

```python
# Esperimento di Buffon

fig4, ax0 = plt.subplots(nrows=1, ncols=1, figsize=(10,10))

n_blocks = 100

mean_prog = np.zeros(n_blocks)
err_prog = np.zeros(n_blocks)
x = np.arange(n_blocks)

expected_pi = [math.pi for i in x] #Valore atteso di pi_greco

file_mean = open("01.3/Data/pi.out")
file_err = open("01.3/Data/error_pi.out")

for i in range(n_blocks):
    mean_prog[i] = float(file_mean.readline())
    err_prog[i] = float(file_err.readline())
    
file_mean.close()
file_err.close()

ax0.errorbar(x,mean_prog,yerr=err_prog)
ax0.plot(x, expected_pi, 'r', label="Valore atteso")

ax0.set_xlabel('# Blocchi', fontsize='large')
ax0.set_ylabel('$\pi$', fontsize='large')
ax0.set_title("Stima di $\pi$ attraverso l'esperimento di Buffon", fontsize='x-large')
ax0.grid(True, linestyle="--", linewidth=0.2, color='.25', zorder=-10)
ax0.legend(fontsize="large")

plt.show()
```

Il valore ricavato dalla simulazione converge al valore atteso all'aumentare del numero di blocchi, entro una deviazione standard.
